```python
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
n = len(nums)
max_i = float("-inf")
for i in range(0, n):
  total = 0
  for j in range(i, n):
    total = total + nums[j]
    max_i = max(max_i, total)
print(max_i)
# This is a Brute force technique here the time complexity is O(N(N+1)/2) ie: O(N^2)
# The space complexity for this piece of code is O(1)
```

```python
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
n = len(nums)
max_i = float("-inf")
total = 0
for i in range(0,n):
  total = max(total + nums[i], nums[i])
  max_i = max(max_i, total)
  if total < 0:
    total = 0
print(max_i)
# The time complexity of this program is O(N)
# The space complxity of this program is O(1)
```
