### Problem Statement: Right Rotating an Array by `q` Positions

Given an array (or list) of elements and a non-negative integer `q`, the task is to rotate the elements of the array to the right by `q` positions.

**What does "right rotation" mean?**

When you right rotate an array by one position, the last element moves to the first position, and all other elements shift one position to the right.

For example, if you have the array `[1, 2, 3, 4, 5]` and you right rotate it by one position, the result is `[5, 1, 2, 3, 4]`.

**Rotating by `q` positions:**

Rotating an array by `q` positions means applying the one-position right rotation process `q` times.

For example, if you have the array `[1, 2, 3, 4, 5]` and you want to right rotate it by `q = 2` positions:

1. **First rotation:** `[5, 1, 2, 3, 4]`
2. **Second rotation:** `[4, 5, 1, 2, 3]`

So, the final result of right rotating `[1, 2, 3, 4, 5]` by 2 positions is `[4, 5, 1, 2, 3]`.

**Key considerations:**

*   **In-place rotation:** Sometimes, the problem asks for the rotation to be done "in-place," meaning you should modify the original array directly without creating a new one.
*   **Handling `q` greater than the array length:** If `q` is greater than the length of the array, rotating by `q` is the same as rotating by `q % n`, where `n` is the length of the array. This is because rotating by `n` positions brings the array back to its original state.

In summary, the problem of right rotating an array by `q` positions involves moving the last `q` elements to the beginning of the array and shifting the remaining elements to the right.

```python
nums = [5,3,2,4,5,6,7,8,2,4,9]
n = len(nums)
temp = nums[n-1]
for i in range(n-2, -1, -1):
  nums[i+1] = nums[i]
nums[0] = temp
print(nums)
# the time complexity of the solution to this problem is O(N-1) ie O(N)
# the space complexity of the solution is O(1)
```
