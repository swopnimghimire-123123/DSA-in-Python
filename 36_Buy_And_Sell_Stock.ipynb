```python
# This is a bruteforce solution
# The time complexity for this piece of code is O(N(N+1)/2) ie: O(N^2)
# The space complexity for this piece of code is O(1)
prices = [7,2,1,5,6,4,8]
max_profit = 0
for i in range (0, len(prices)):
  for j in range(i+1, len(prices)):
    if prices[j] > prices[i]:
      profit = prices[j] - prices[i]
      if profit > max_profit:
        max_profit = profit
print(max_profit)
```

```python
# This is a optimal piece of code
# The time complexity for this piece of code is O(N)
# The space complexity for this piece of code is O(1)
import builtins

prices = [7,2,1,5,6,4,8]
optimal_profit = 0
min_price = float('inf')
n = len(prices)

for i in range (0,n):
  min_price = builtins.min(min_price, prices[i])
  optimal_profit = builtins.max(optimal_profit, prices[i] - min_price)
print(optimal_profit)
```

We imported builtins because there was a variable in the environment named max which was interfering with the built-in max() function. By importing builtins, we can explicitly refer to the built-in max() and min() functions as builtins.max() and builtins.min() to avoid this conflict and ensure we are using the correct functions.
