### DSA Perspective: Time and Space Complexity

Let's analyze the time and space complexity of the methods we used to find the largest element in an array.

**Method 1: Iterating from scratch**

*   **Time Complexity:** O(n), where 'n' is the number of elements in the array. We need to iterate through the entire array once to find the largest element.
*   **Space Complexity:** O(1). We only use a constant amount of extra space to store the `largest_element_from_scratch` variable.

**Method 2: Using `max()` function**

*   **Time Complexity:** O(n) in the worst case. The `max()` function in Python iterates through the elements of the iterable to find the maximum.
*   **Space Complexity:** O(1). Similar to the iterative method, the `max()` function generally uses constant extra space.

**Method 3: Sorting the array**

*   **Time Complexity:** O(n log n) for most efficient sorting algorithms (like Timsort used in Python). Sorting takes the dominant time.
*   **Space Complexity:** O(n) in the worst case for some sorting algorithms (like Timsort) due to the need for temporary storage during sorting.

```python
my_array = [10, 4, 25, 7, 15]
my_array.sort()
largest_element = my_array[-1]
print(largest_element) # Output: 25
```

```python
my_array = [10, 4, 25, 7, 15]
largest_element = max(my_array)
print(largest_element)
```

```python
my_array = [10, 4, 25, 7, 15]
largest_element_from_scratch = my_array[0]

for element in my_array:
  if element > largest_element_from_scratch:
    largest_element_from_scratch = element

print(largest_element_from_scratch)
```
