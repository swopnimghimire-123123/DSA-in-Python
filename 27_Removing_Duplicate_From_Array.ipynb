```python
# BRUTE FORCE
nums = [1,2,3,4,5,5,6,7,7]
filtered_nums = []
n = len(nums)
freq_map = {}
for i in range(0,n):
  freq_map[nums[i]] = 0

j = 0
for k in freq_map :
  filtered_nums.append(k)
  j += 1
print(filtered_nums)
# The time complextiy is O(2N)
# The space complexity at worse case O(N)
```

```python
# OPTIMAL SOLUTION
nums = [1,2,3,4,5,5,6,7,7]
n = len(nums)
if n == 1:
  print(nums)
i = 0
j = i + 1
while j < n:
  if nums[j] != nums[i]:
    i += 1
    nums[i],nums[j] = nums[j], nums[i]
  j += 1
print(nums[:i+1])
# The time complexity is O(N)
# The space complexity is O(1)
```
