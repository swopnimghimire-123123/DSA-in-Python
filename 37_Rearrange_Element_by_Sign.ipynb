```python
# This is a brute force method
# The time complexity for this piece of code is 0(N)+O(N/2)
# The space complexity for this piece of code id 0(N/2+N/2)
nums = [5,10,-3,-1,-10,6]
positive = []
negative = []
for i in nums:
  if i >= 0:
    positive.append(i)
  else:
    negative.append(i)
for i in range(0, len(positive)):
  nums[2*i] = positive[i]
  nums[(2*i)+1] = negative[i]
print(nums)
```

```python
# This is a optimal solution
# The time complexity for this one is 0(N) + 0(N)
# The space complexity for this one is 0(1)/ 0(N)
nums = [5,10,-3,-1,-10,6]
n = len(nums)
result = [0] * len(nums) # It's tc is 0(N)
# the spaces complexity is 0(1) but it can be ignored
pos_index, neg_index = 0, 1
for i in range(0, n):
  if nums[i] >= 0 : # It's tc is 0(N)
    result[pos_index] = nums[i]
    pos_index += 2
  else:
    result[neg_index] = nums[i]
    neg_index += 2
print(result)
```
