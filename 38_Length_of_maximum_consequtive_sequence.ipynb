```python
# This is a brute force method
# It's time complexity is 0(N^2)
# It's space complexity is 0(1) although it takes space for 4 variable we neglict that.
nums = [1,99,101,98,2,5,3,100]
n = len(nums)
max_count = 0
for i in range(0, n): # it's time complexity is 0(N)
  num = nums[i]
  count = 1
  while num+1 in nums: # it's time complexity is 0(N)
    num = num+1
    count = count+1
  max_count = max(max_count, count)
print(max_count)
```

```python
# This is a better solution
# The time complexity for this code is O(NlogN + N)
# The space complxity for this code is O(1)
nums = [1,99,101,98,2,5,3,100]
nums.sort() # IT'S TC IS O(NlogN)
count = 0
last_smaller = float("-inf")
longest = 0
for i in range(0, n): # IT'S TIME COMPLEXITY IS O(N)
  num = nums[i]
  if num-1 == last_smaller:
    count += 1
    last_smaller = num
  elif num != last_smaller:
    count = 1
    last_smaller = num
  longest = max(longest, count)
print(longest)
```

```python
# The time complexity for this code is 0(N+N+N) ie 0(3N) ie 0(N)
# The space complexity fot this code is 0(N)
# This is the optimal solution

nums = [1,1,99,101,98,2,5,3,100]
my_set = set()
n = len(nums)
for i in range(0, n):
  my_set.add(nums[i])
largest = 0
for num in my_set:
  if num - 1 not in my_set:
    x = num
    count = 1
    while x+1 in my_set:
      count += 1
      x += 1
    largest = max(largest, count)
print(largest)
```
