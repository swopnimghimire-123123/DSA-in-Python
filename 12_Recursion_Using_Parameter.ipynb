# Topic: Parameterized Recursion in Python

## Introduction
**Parameterized recursion** is a type of recursion where **additional parameters** are passed to the recursive function to **carry information** or **accumulate results**.  

Advantages:  
- Reduces the need for global variables  
- Helps maintain intermediate results  
- Often more efficient than returning values and combining them  

---

## Example 1: Factorial Using Parameterized Recursion

### Question
Compute the factorial of a number `n` using a helper parameter to carry the result.  
- Input: `5` → Output: `120`

### Approach
1. Use a helper function `fact_helper(n, acc)` where `acc` accumulates the result.  
2. Base case: `n = 0` → return `acc`  
3. Recursive case: call `fact_helper(n-1, acc*n)`  

### Complexity Analysis
| Operation | Complexity |
|-----------|------------|
| Time      | O(n)       |
| Space     | O(n) (call stack) |

### Edge Cases
- `n = 0` → 1  
- Negative numbers → not allowed for factorial


```python
def factorial_param(n, acc=1):
    if n == 0:
        return acc
    return factorial_param(n-1, acc * n)

print("Factorial of 5:", factorial_param(5))
```

## Example 2: Sum of Digits Using Parameterized Recursion

### Question
Find the sum of digits of a number `n` using a parameter to accumulate the sum.  
- Input: `1234` → Output: `10`

### Approach
1. Use a helper parameter `acc` to accumulate the sum of digits.  
2. Base case: `n = 0` → return `acc`  
3. Recursive case: `acc + sum_of_digits_param(n // 10, acc + n % 10)`  

### Complexity Analysis
| Operation | Complexity |
|-----------|------------|
| Time      | O(d) → O(log n) |
| Space     | O(d) → O(log n) |

### Edge Cases
- `n = 0` → 0  
- Negative numbers → use absolute value


```python
#  Sum of digits using parameterized recursion
def sum_of_digits_param(n, acc=0):
    if n == 0:
        return acc
    return sum_of_digits_param(n // 10, acc + n % 10)

print("Sum of digits of 1234:", sum_of_digits_param(1234))
```
