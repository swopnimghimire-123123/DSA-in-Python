```python
nums = [1,0,3,4,3,0,0,3,5,1]
zeros = []
numbers = []
for items in nums:
  if items == 0 :
    zeros.append(items)
  else:
    numbers.append(items)
solved_nums = numbers + zeros
print(nums)
print(solved_nums)
# Time Complexity: O(n) - The code iterates through the list 'nums' once.
# Space Complexity: O(n) - The code creates two new lists, 'zeros' and 'numbers', in the worst case, they can store up to n elements.
```

```python
nums = [1,0,3,4,3,0,0,3,5,1]
n = len(nums)
temp = []
for i in range(0,n):
  if nums[i] != 0 :
    temp.append(nums[i])
n2 = len(temp)
for i in range(0,n2):
  nums[i] = temp[i]
for i in range(n2, n):
  nums[i] = 0
print(nums)
# the time complexity of this code is O(N+N) ie O(N)
# the space complexity of this code is O(N)
```

```python
nums = [1,0,3,4,3,0,0,3,5,1]

pos = 0
for i in range(len(nums)):
    if nums[i] != 0:
        nums[pos], nums[i] = nums[i], nums[pos]
        pos += 1
# nums is modified in place
print(nums)  # Output: [1, 3, 4, 3, 3, 5, 1, 0, 0, 0]
# The time complexity of this piece of code is O(N)
# The space complxity of this piece of code is O(1)
```
