```python
nums_1 = [1,0,3,4]
nums_2 = [1,2,3,4,5,6,8,9]
nums_3 = [3,0,1]
def finding_missing_number(number):
  n = len(number)
  for i in range(n): # it's time compelxity is O(N)
    if i not in number : # it's time complexity is also O(N)
      print(i)
finding_missing_number(nums_1)
print("\n")
finding_missing_number(nums_2)
print("\n")
finding_missing_number(nums_3)
# the total time complexity is 0(N^2)
# the space complxity of this code is O(1)
```

```python
nums_1 = [1,0,3,4]
nums_2 = [1,2,3,4,5,6,8,9]
nums_3 = [3,0,1]

def finding_missing_number(number):
  freq = {}
  n = len(number)
  for i in range(0, n+1): # this has time complexity of O(N)
    freq[i] = 0
  for num in number: # this has a time complxity of O(N)
    freq[num] = 1
  for k,u in freq.items(): # this has a time complxity of O(N)
    if u == 0:
      print(k)

print(finding_missing_number(nums_1))
print(finding_missing_number(nums_2))
print(finding_missing_number(nums_3))
# The total time complxity of the program is O(3N) this is approximately O(N)
# The space complxity of this program is O(N)
```
