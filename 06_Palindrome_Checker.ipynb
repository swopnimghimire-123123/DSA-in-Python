# Problem: Check if an Integer is a Palindrome (Without Converting to String)

## Question
Given an integer `n`, check if it is a **palindrome** without converting it into a string.  
A number is a palindrome if it reads the same forwards and backwards.  

Example:  
- Input: `3456` → Output: `not palindrome`  
- Input: `1221` → Output: `palindrome`

---

## Explanation
We can reverse the number mathematically without converting it to a string:  

1. Initialize `result = 0`.  
2. Extract digits from `nums` using `% 10`.  
3. Construct the reversed number: `result = result * 10 + last_digit`.  
4. Remove the last digit from `nums` using integer division (`// 10`).  
5. After the loop, compare `result` with the original number `n`.  

---

## Approach
1. Copy the number into a temporary variable `nums`.  
2. Initialize `result = 0`.  
3. While `nums != 0`:  
   - Extract last digit using `% 10`  
   - Update reversed number: `result = result * 10 + last_digit`  
   - Remove last digit from `nums` using `// 10`  
4. Compare `result` with `n`.  
5. If equal → palindrome, else → not palindrome.

---

## Complexity Analysis

| Operation                | Complexity          |
|--------------------------|------------------|
| Time                     | O(d) → O(log n)  |
| Space                    | O(1)             |

**Explanation:**  
- `d` = number of digits in `n`  
- Each iteration processes one digit → O(d)  
- We only store `result` and a few variables → O(1)  

---

## Edge Cases
- `n = 0` → palindrome  
- Negative numbers → generally **not considered palindromes** (can use absolute value if needed)  
- Single-digit numbers → always palindromes


```python
n = 3456
nums = n
result = 0

while nums != 0:
    last_digit = nums % 10
    result = result * 10 + last_digit
    nums = nums // 10

if result == n:
    print("palindrome")
else:
    print("not palindrome")
```
