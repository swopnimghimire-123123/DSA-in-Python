## Problem Statement: Finding Indices of Two Numbers with a Target Sum

Given a list of numbers and a target sum, find the indices of two numbers in the list that add up to the target sum.

**Assumptions:**

* There is at least one pair of numbers in the list that sum up to the target.
* If multiple pairs sum up to the target, return the indices of the first pair found.

**Input:**

* A list of numbers (e.g., `[2, 7, 11, 15]`)
* A target sum (e.g., `9`)

**Output:**

* A list containing the indices of the two numbers that sum up to the target (e.g., `[0, 1]`, since `list[0] + list[1] = 2 + 7 = 9`)

```python
nums = [1,2,3,4,5,6,7,8,9]
target = 17
n = len(nums)
for i in range(0, n-1):
  for j in range(i+1, n):
    if nums[i] + nums[j] == target:
      print([i, j])
      break
  else:
    continue
  break
# This is a brute force solution
# The time complexity for this problem is O(N(N+1)/2) ie O(N^2)
# The space complexity for this problem is O(1)
```

```python
# OPTIMAL SOLUTION
nums = [1,2,3,4,5,6,7,8,9]
target = 17
hash_map = {}
n = len(nums)
for i in range(0, n): # The time complexity for this one is O(N)
  remaining = target - nums[i]  # O(!)
  if remaining in hash_map : # O(1)
    print((hash_map[remaining]),i)
  hash_map[nums[i]] = i
# The total time complexity of this program is O(N)
# The space complexity fo this program is also O(N)
```
